Hierarchical clustering of transcription data 
===
This notebook applies hierarchical GPy clustering algorithm to cluster 2000 transcription data. Hierarchical GPy clustering algorithm was developed by Dr James Hensman from The University of Sheffield (http://staffwww.dcs.sheffield.ac.uk/people/J.Hensman/) . 

In this example, we would like to cluster the dataset for control and perturbed conditions together, so we will use 'IndependOutputs' kernel from GPy for this purpose.

### Example

Import packages for later usage

```python
import numpy as np
import pylab as pb
import GPy
from GPy import plotting
from matplotlib import pyplot as plt
from colvb import MOHGP
%matplotlib inline
```

Firstly, let us import the time course data and normalize the data for later usage

```python
###  import data for later application, where the first column is the probe ID, 1-8 columns of the data are the data for 
###  control condition and 9-16 columns of the data are the data for perturbed condition. 
M = 1000  ### only import the first 2 data 

data = np.genfromtxt('http://personalpages.manchester.ac.uk/staff/jing.yang/Data/top2000_data.txt', dtype=np.str, usecols=range(15), skip_header=0)[:M]
Y = data[:,1:15]
Y.dtype = np.float64
### use logscaled data
Y = np.log(Y)
### data normalization 
Y -= Y.mean(1)[:,None]
Y /= Y.std(1)[:,None]

### seperate original data to form f and g
f = Y[:,range(7)]
g = Y[:,range(7,14)]
### set probe ID
probe_ID = data[:,0]
###  Experiment time points are saved in 'times'
times= np.array([0,3.5,7.0,10.5,14.0,17.5,21.0])

```

Now we are going to build a hierarchical kernel for the data and its replicate. In this case, we only have one replicate, so the replicate model may not be used. But we list it here for future usage. We use IndependentOutputs kernel from GPy as the kernel for time axis.  

```python
kern_upper = GPy.kern.Matern32(input_dim=1, variance=1.0, lengthscale=12.0, active_dims=[0], name='upper')
kern_lower = GPy.kern.Matern32(input_dim=1, variance=0.1, lengthscale=20.0, active_dims=[0], name='lower') + GPy.kern.White(1,0.01, active_dims=[0])
k_hierarchy = GPy.kern.Hierarchical(kernels=[kern_upper, kern_lower])
k = GPy.kern.IndependentOutputs(GPy.kern.RBF(1))
X = np.r_[np.c_[times, np.zeros_like(times)], np.c_[times, np.zeros_like(times)]]  ### one replciate only for control and perturbed conditions
alpha =0.1 ### DP concentration
Nclust = 20 ### initial number of clusters
m = MOHGP(X, k, k_hierarchy, Y, K = Nclust, prior_Z='DP', alpha = alpha )
print(X)
print(Y.shape)
```

```python
m.plot(1,1,1,0,0,0, fixed_inputs=[(1,0)])
```

```python
m.optimize()
```

```python
m.plot(1,1,1,0,0,0, fixed_inputs=[(1,0)])
```
